## Load the Data

```python
import pandas as pd
```

```python
bigmart= pd.read_csv('train_bm.csv')
bigmart.head()
```

```python
bigmart['Item_Type'].value_counts()
```

## Frrequency Encoding

```python
temp = bigmart['Item_Type'].value_counts()
temp = pd.DataFrame(temp)
temp.head()
```

```python
temp['Item_Type_count'] = temp['Item_Type']
temp['Item_Type'] = temp.index
temp.head()
```

```python
bigmart = pd.merge(bigmart, temp, on='Item_Type', how='left')
bigmart[['Item_Identifier', 'Item_Type', 'Item_Type_count']].head()
```

```python
bigmart[['Item_Identifier', 'Item_Type', 'Item_Type_count']].head(10)
```
